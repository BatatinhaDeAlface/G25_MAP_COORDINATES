```python
rg['centroid'] = rg.geometry.centroid
countries['centroid'] = countries.geometry.centroid

points_nuts2 = pd.DataFrame({
    'lon': rg.centroid.x,
    'lat': rg.centroid.y,
    'Distance': rg.Distance
})

points_country = pd.DataFrame({
    'lon': countries.centroid.x,
    'lat': countries.centroid.y,
    'Distance': countries.Distance
})

points = pd.concat([points_nuts2, points_country]).dropna()

lon_min, lon_max = -25, 40
lat_min, lat_max = 25, 72

grid_lon, grid_lat = np.meshgrid(
    np.linspace(lon_min, lon_max, 600),
    np.linspace(lat_min, lat_max, 400)
)

grid_distance = griddata(
    points[['lon', 'lat']],
    points['Distance'],
    (grid_lon, grid_lat),
    method='linear'
)


def maximize_contrast(x, gamma , log_factor):
    x_norm = (x - np.nanmin(x)) / (np.nanmax(x) - np.nanmin(x))
    x_log = np.log1p(x_norm * log_factor)
    x_log /= np.nanmax(x_log)
    x_contrast = np.power(x_log, gamma)
    return x_contrast
grid_contrast = maximize_contrast(grid_distance,1.5, 600)

colormaps = ["Blues_r", "viridis_r", "magma_r", "afmhot_r", "copper_r"]

def plot_interpolated_map(cmap="Blues_r"):
    fig, ax = plt.subplots(figsize=(12, 12))

    im = ax.imshow(grid_contrast, origin='lower',
                   extent=(lon_min, lon_max, lat_min, lat_max),
                   cmap=cmap, alpha=0.8)

    countries.boundary.plot(ax=ax, color='black', linewidth=0.5)

    ax.set_xlim(lon_min, lon_max)
    ax.set_ylim(lat_min, lat_max)

    plt.colorbar(im, ax=ax)

    plt.tight_layout()
    plt.show()

@interact(cmap=colormaps)
def update_interpolated(cmap):
    plot_interpolated_map(cmap)
```
